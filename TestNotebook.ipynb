# Compute metrics

```python
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval
```

```python
test = 'dataset/test_set/test.json'

inputs = COCO(test)
inputs.cats
```

```python
detections = inputs.loadRes('247195_242765.json')

detections.cats
```

```python
# running evaluation
annType = 'bbox'
cocoEval = COCOeval(inputs, detections, annType)
#cocoEval.params.imgIds  = imgIds
cocoEval.evaluate()
cocoEval.accumulate()
cocoEval.summarize()
```

```python
# running evaluation
annType = 'segm'
cocoEval = COCOeval(inputs, detections, annType)
#cocoEval.params.imgIds  = imgIds
cocoEval.evaluate()
cocoEval.accumulate()
cocoEval.summarize()
```
